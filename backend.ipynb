```python
import streamlit as st
import pandas as pd
import os

# =====================================
# Streamlit Config
# =====================================
st.set_page_config(page_title="News Search Engine", layout="wide")

st.title("ğŸ“¡ News Search Engine (Gold Layer)")
st.write("Search by date, stock symbol, sentiment or keywords.")

# =====================================
# 1) è‡ªåŠ¨åŠ è½½ CSV
# =====================================
DEFAULT_PATH = r"D:\Columbia\Fall2025\5400\project\silver_with_sentiment_sample.csv"

if not os.path.exists(DEFAULT_PATH):
    st.error(f"âŒ æ‰¾ä¸åˆ°æ–‡ä»¶ï¼š{DEFAULT_PATH}")
    st.stop()

df = pd.read_csv(DEFAULT_PATH)
st.success(f"CSV å·²åŠ è½½ï¼å…±æœ‰ {df.shape[0]} è¡Œï¼Œ{df.shape[1]} åˆ—")

# =====================================
# 2) æœç´¢æ¡ä»¶åŒºåŸŸ
# =====================================

st.sidebar.header("ğŸ” æœç´¢æ¡ä»¶")

# ---- æ—¥æœŸæœç´¢ ----
date_min = df["Date"].min() if "Date" in df.columns else None
date_max = df["Date"].max() if "Date" in df.columns else None

selected_date = st.sidebar.date_input(
    "æŒ‰æ—¥æœŸè¿‡æ»¤ï¼ˆå¯é€‰ï¼‰",
    value=None
)

# ---- è‚¡ç¥¨ä»£ç æœç´¢ ----
symbols = ["(å…¨éƒ¨)"] + sorted(df["Stock_symbol"].dropna().unique().tolist())
symbol = st.sidebar.selectbox("æŒ‰è‚¡ç¥¨ä»£ç æœç´¢", symbols)

# ---- sentiment æœç´¢ ----
if "sentiment_score_signed" in df.columns:
    sentiment_range = st.sidebar.slider(
        "Sentiment Score èŒƒå›´",
        float(df["sentiment_score_signed"].min()),
        float(df["sentiment_score_signed"].max()),
        (float(df["sentiment_score_signed"].min()),
         float(df["sentiment_score_signed"].max()))
    )
else:
    sentiment_range = None

# ---- keyword æœç´¢ ----
keyword = st.sidebar.text_input("æ ‡é¢˜å…³é”®å­— (å¯é€‰)")

# =====================================
# 3) æ‰§è¡Œè¿‡æ»¤
# =====================================
filtered = df.copy()

if selected_date:
    filtered = filtered[ filtered["Date"] == str(selected_date) ]

if symbol != "(å…¨éƒ¨)":
    filtered = filtered[ filtered["Stock_symbol"] == symbol ]

if sentiment_range:
    filtered = filtered[
        (filtered["sentiment_score_signed"] >= sentiment_range[0]) &
        (filtered["sentiment_score_signed"] <= sentiment_range[1])
    ]

if keyword:
    filtered = filtered[ filtered["Article_title"].str.contains(keyword, case=False, na=False) ]

st.subheader(f"ğŸ” æœç´¢ç»“æœï¼ˆ{filtered.shape[0]} æ¡ï¼‰")
st.dataframe(filtered.head(100))

# =====================================
# 4) é€‰æ‹©æ–‡ç« 
# =====================================
st.markdown("---")
st.subheader("ğŸ“° æŸ¥çœ‹æ–‡ç« è¯¦æƒ…")

if filtered.shape[0] == 0:
    st.warning("æ²¡æœ‰ç»“æœï¼Œè¯·è°ƒæ•´æœç´¢æ¡ä»¶ï¼")
    st.stop()

selected_id = st.selectbox("é€‰æ‹©æ–°é—» ID", filtered["news_id"].unique())
row = filtered[filtered["news_id"] == selected_id].iloc[0]

# =====================================
# 5) æ˜¾ç¤ºå®Œæ•´æ–‡ç« 
# =====================================
st.markdown("## ğŸ“ Full Article")

if "Article" in df.columns:
    st.markdown(row["Article"], unsafe_allow_html=True)
else:
    st.write("âŒ æ­¤ CSV æ²¡æœ‰ Article å­—æ®µ")


# =====================================
# 6) Summaries
# =====================================
st.markdown("---")
st.markdown("## ğŸ§  Summaries")

summary_fields = [
    "Lsa_summary",
    "Luhn_summary",
    "Textrank_summary",
    "Lexrank_summary"
]

for col in summary_fields:
    if col in df.columns:
        st.markdown(f"### {col}")
        st.markdown(row[col], unsafe_allow_html=True)

# =====================================
# 7) Metadata
# =====================================
st.markdown("---")
st.markdown("## ğŸ“Œ Metadata")

for col in ["Date", "Publisher", "Author", "Stock_symbol", "Url", "sentiment_label", "sentiment_score_signed"]:
    if col in df.columns:
        st.write(f"**{col}:** {row[col]}")
```
